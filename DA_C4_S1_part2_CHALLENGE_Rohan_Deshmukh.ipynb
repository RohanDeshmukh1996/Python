```python
import numpy as np
data=np.genfromtxt('DS1_C4_S1_Shopping_Data_Challenge.csv', delimiter=',', skip_header=1)
```

```python
# Task 1
# Giving condition by where to specific column and row where value is greater than 80 to get index
x=np.where(data[0:,3]>80)
print("%10s %8s %22s %25s"%("CustomerID","Age","Annual_Income_(k$)","Spending_Score(1-100)"))
# using index to get values of row and column and storing it in object
a=data[x][0:,0]
b=data[x][0:,1]
c=data[x][0:,2]
d=data[x][0:,3]
for i in range(a.size):
    print("%6s %11s %13s %25s"%(int(a[i]),int(b[i]),c[i],d[i]))
```

```python
# Task 2
# Giving condition by where to specific column and row where value is between 20 and 24 to get index
x=np.where((data[0:,1]>=20)&(data[0:,1]<24))
# using index to get values of row and column and storing it in object
a=data[x][0:,0]
b=data[x][0:,1]
c=data[x][0:,2]
d=data[x][0:,3]
print("%10s %8s %22s %25s"%("CustomerID","Age","Annual_Income_(k$)","Spending_Score(1-100)"))
for i in range(a.size):
    print("%6s %11s %13s %25s"%(int(a[i]),int(b[i]),c[i],d[i]))
```

```python
# Task 3
# getting min and max value from a specific column and storing it in object 
a=data[:,2].max()
b=data[:,2].min()
# using value from above to get index of that specific value
x=np.where(data[0:,2]==a)
y=np.where(data[0:,2]==b)
# using index from above to get complete detail from each column
Max=data[x][0:,0]
Min=data[y][0:,0]
print("customer id with lowest annual income")
for i in range(Min.size):
    print(int(Min[i]))
print("customer id with highest annual income")
for i in range(Max.size):
    print(int(Max[i]))
```

```python
# Task 4
# using index to get values of row and column and storing it in object less than 45 and greater than 35 & 75 to get index
x=np.where((data[0:,1]<45)&(data[0:,2]>35)&(data[0:,3]>75))
# using index to get values of row and column and storing it in object 
a=data[x][0:,0]
b=data[x][0:,1]
c=data[x][0:,2]
d=data[x][0:,3]
print("Amount of customer meeting that criteria are :",a.size)
print("%10s %8s %22s %25s"%("CustomerID","Age","Annual_Income_(k$)","Spending_Score(1-100)"))
for i in range(a.size):
    print("%6s %11s %13s %25s"%(int(a[i]),int(b[i]),c[i],d[i]))
```
