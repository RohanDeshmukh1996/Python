```python
# Displaying order items quantity and cost
def display(d,items): 
    print("%43s"%("ordered Items\n"))
    print("%10s %15s %24s %10s" %("Sr.no", "item", "quantity","Cost in $"))
    for i in items:
        print("%8s %28s %10s %10s" %(i,items[i]["item"],d[i],(d[i]*items[i]["Cost"])))

        
        
# for getting sum of values        
def total(d,items):
    sum=0
    for i in items:
        sum=sum+(d[i]*items[i]["Cost"])
    return sum


# Charging the customer based on miles
def charge(a):
    if a>=15:
        return 20
    elif a>=10:
        return 14
    elif a>=5:
        return 12
    else:
        return 10
 
# nested dictionary of items
items = {
    1: {
        "item": "Book a4_100pg pack of 4",
        "quantity": 50,
        "Cost": 4.5,
    },
    2: {
        "item": "Book a4_100pg pack of 8",
        "quantity": 40,
        "Cost": 8.0,
    },
    3: {
        "item": "Book a5_100pg pack of 4",
        "quantity": 70,
        "Cost": 5.0,
    },
    4: {
        "item": "Book a5_100pg pack of 8",
        "quantity": 30,
        "Cost": 9.0,
    },
    5: {
        "item": "Book a4_160pg pack of 2",
        "quantity": 80,
        "Cost": 4.2,
    },
    6: {
        "item": "Book a5_160pg pack of 2",
        "quantity": 80,
        "Cost": 5.2,
    }
}

condn="yes" # condition for while loop to work

while(condn=="yes"):

    print("%43s"%("Available Items\n"))
    print("%10s %15s %24s %10s" %("Sr.no", "item", "quantity","Cost in $"))#using %*s method to show spacing
    for i in items:
        print("%8s %28s %10s %10s" %(i,items[i]["item"],items[i]["quantity"],items[i]["Cost"]))
    
    d={}# for storing input
    for i in items:
        print("How many",items[i]["item"],"do you want:")# asking for values
        n=int(input())
        # conditioning value n to provide validate quantity in store 
        if (items[i]["quantity"])>0 and (items[i]["quantity"])>=n: 
            d.update({i:n})
        elif (items[i]["quantity"])>0 and (items[i]["quantity"])<n:
            print("quantity not available")
            print("Enter proper quantity ( less than",(items[i]["quantity"]),")")
        else:
            print("item is out of stock")
            d.update({i:0})
     
    # updating quantity in main dataset
    for i in items:
        if(d[i])!=0:
            items[i]["quantity"]=(items[i]["quantity"])-d[i]
        
        
    # getting customer data for delivery   
    print("\nDelivery details")
    name=input("\n Enter Name : ")
    add=input("Enter address : ")
    dist=int(input("Distance from store in miles :"))
    dc=charge(dist)# calling function charge and receiving output
    
    
    display(d,items) # calling function display   
    sum=total(d,items) #calling funtion total
    TotalCharges=sum+dc
    
    print("\n\nTotal item cost :",sum)
    print("Total Bill amount = total item cost + dilevery charges:",TotalCharges)
    print("\n\nName:",name)
    print("Address:",add)
    abc=input("\n\n Do you want to continue(yes or no) :")#asking customer if he/she want to continue
    condn=abc.lower() # making abc value lower to match condition statement
```
