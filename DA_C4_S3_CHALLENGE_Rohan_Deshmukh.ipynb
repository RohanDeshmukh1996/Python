```python
# Task 1
# using read_csv to convert csv data to dataframe as data
data=pd.read_csv("DS1_C4_S3_Loan_Data_Challenge.csv")
data.info()
```

```python
# Task 2
# giving condition to data for(Gender=female,self employed,income>4000$)
data[(data['Gender']=='Female')&(data['Self_Employed']=='Yes')&(data['ApplicantIncome']>4000)]
```

```python
# Task 3
#giving condition to data for(Gender=male,graduate,loan status=y)
a= data[(data['Gender']=='Male')&(data['Education']=='Graduate')&(data['Loan_Status']=='Y')]
# using len function on new dataframe to get length of it
print("Number of customer who is Male, Graduate and loan status in yes : ",len(a))
a
```

```python
# Task 4
print("Number of Customer who loan status is Y, has more than 3 dependant and have proprty in rural area :")
#giving condition to data for(Gender=male,graduate,loan status=y)
print(len(data[(data['Loan_Status']=='Y')&(data['Dependents']=='3+')&(data['Property_Area']=='Rural')]))
```

```python
# Task 5
#giving condition to data for ( populate=urban, graduate, loan amount>=290 )
d=data[(data['LoanAmount']>=290)&(data['Education']=='Graduate')&(data['Property_Area']=='Urban')]
a=d.shape[0]
#giving condition to data for ( populate=urban,graduate,loan status=y )
b=data[(data['LoanAmount']>=290)&(data['Education']=='Graduate')&(data['Property_Area']=='Urban')&(data['Loan_Status']=='Y')].shape[0]
# getting value in %
c=(b/a)*100
print("Approval of loan for a certain person with similar conditions are ",c,"%")
d.Loan_Status.value_counts()
```

```python
# Task 6
#giving condition to data for ( married=yes, Selfemployed, loan Status=N ,populate=urban)
a=data[(data['Married']=='No')&(data['Self_Employed']=='Yes')&(data['Property_Area']!='Urban')&(data['Loan_Status']=='N')]
# using count function for getting count of gender column
a.Gender.value_counts()
```

```python
# Task 7
#giving condition to data for (populate=urban) and taking average 
avg_u=data[data['Property_Area']=='Urban'].ApplicantIncome.mean()
#giving condition to data for (populate=semiurban) and taking average 
avg_su=data[data['Property_Area']=='Semiurban'].ApplicantIncome.mean()
#giving condition to data for (populate=rural) and taking average 
avg_r=data[data['Property_Area']=='Rural'].ApplicantIncome.mean()
# using rounf function to get only 2  digit after decimal
print('Average income of customer who has property in Urban :',round(avg_u,2))
print('Average income of customer who has property in Semiurban :',round(avg_su,2))
print('Average income of customer who has property in Rural :',round(avg_r,2))
```

```python
# Task 8
#giving condition to data for ( Loan Status=N, income>4000, coapplicantincome>1000 ,populate!=urban)
a=data[(data['Property_Area']!='Urban')&(data['ApplicantIncome']>4000)&(data['CoapplicantIncome']>1000)
       &(data['Loan_Status']=='N')]
# displaying no. of customer  for above condition
print("Number of customer who dosen't have Urban property, Income is more than $4000 and Coapplicant Income is grater than $1000 and Loan Status is No : ",len(a))
print("--------------------------------------------------------------------------------------------------------")
# overwriting the  value where loan status= N to Y using loc function   
a.loc[a.Loan_Status=='N','Loan_Status']='Y'

# making changes in  original data by overwriting 
data[(data['Property_Area']!='Urban')&(data['ApplicantIncome']>4000)&(data['CoapplicantIncome']>1000)
       &(data['Loan_Status']=='N')]=a
data
```

```python
# Task 8 part 2
print("after making changes")
# reapplying the above code for checking if we made changes to original data
a=data[(data['Property_Area']!='Urban')&(data['ApplicantIncome']>4000)&(data['CoapplicantIncome']>1000)
       &(data['Loan_Status']=='N')]
print("Number of customer who dosen't have Urban property, Income is more than $4000 and Coapplicant Income is grater than $1000 and Loan Status is No : ",len(a))
```
