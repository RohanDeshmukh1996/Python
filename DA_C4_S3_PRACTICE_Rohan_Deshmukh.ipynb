```python
import pandas as pd
import numpy as np
data=pd.read_csv("DS1_C4_S3_SalesData_Practice.csv")
data
```

```python
# Task 1
# using sorting function to sort dataframe by ascending order using particular column 
data.sort_values('Sales',ascending= True)
```

```python
# Task 2
# giving condition to sales that is graeter tha equal to 45000
data[data.Sales>=45000]
```

```python
# Task 3 part 1
# using sorting function to sort dataframe by ascending order using particular column with specific name
california=data[data.State=="California"].sort_values('Sales',ascending= True)
california
```

```python
# Task 3 part 2
# using sorting function to sort dataframe by ascending order using particular column with specific name
texas=data[data.State=="Texas"].sort_values('Sales',ascending= True)
texas
```

```python
# Task 3 part 3
# using sorting function to sort dataframe by ascending order using particular column with specific name
montana=data[data.State=="Montana"].sort_values('Sales',ascending= True)
montana
```

```python
# Task 4
#using sum function to sum of value of particular column
a=california["Sales"].sum()
b=texas["Sales"].sum()
c=montana["Sales"].sum()
print("Total Sales in the State of California",a)
print("Total Sales in the State of Texas",b)
print("Total Sales in the State of Montana",c)
```

```python
# Task 5
#gettin max value of a particular column
a=california[california.Sales==california.Sales.max()]
b=california[california.Sales==california.Sales.min()]
print("Highest sales month in California :")
print(a["Month"])
print("----------------------------------------------------")
print("Lowest sales month in California :")
print(b["Month"])
```

```python
# Task 5 alternative
a = data.loc[(data['State'] == 'California')].max()
print("Highest sales month in California :",a.Month)
b = data.loc[(data['State'] == 'California')].min()
print("----------------------------------------------------")
print("Lowest sales month in California :",b.Month)
```

```python
# Task 6
# Creating new dataframe and storing the sum of value of original dataframe in new one
x={"State":['Alaska', 'Texas', 'California', 'Montana','New Mexico'],
   "Sum_Of_Sales":[data[data.State=='Alaska']['Sales'].sum(),
                   data[data.State=='Texas']['Sales'].sum(),
                   data[data.State=='California']['Sales'].sum(),
                   data[data.State=='Montana']['Sales'].sum(),
                   data[data.State=='New Mexico']['Sales'].sum()]}
New=pd.DataFrame(x)
New
```

```python
#Task7
# creating csv file using python
New.to_csv("total_sales.csv")
```
