```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  
```

```python
data=pd.read_excel('DS1_C4_S5_Employee_Data_Practice.xlsx')
data
```

```python
# Task 1
# using groupby on department and sum on salary for values
x=data.groupby("Department")["Annual Salary ($)"].sum()

l1=list(x.index)
l2=list(x.values)
#using plt.bar to plot bar graph of the list
plt.bar(l1,l2)
```

```python
# Task 2
# using crosstable on department and gender 
x=pd.crosstab(data.Department,data.Gender)
#using plot.bar to plot bar graph of the list
x.plot.bar()
plt.show()
```

```python
# Task 3
#giving condition to data frame 
x=data[(data.Work_Experience==0)&(data.Department=="IT ")]
# displaying mean salary
print("Average Salary of the non Experienced Candidate can be ",x["Annual Salary ($)"].mean())
```

```python
# Task 4
# ploting pie chart and using autopct to show percentage
plt.pie(l2,labels=l1,autopct='%.2f %%')
plt.show()
```

```python
# Task 5
# ploting histogram base on age and 5 bins
plt.hist(data['Age'],ec="r",bins=5)
plt.show()
```

```python
# Task 6
x=data["Work_Experience"]
y=data["Annual Salary ($)"]
# ploting scatter plot with trend line
plt.scatter(x,y,color="white",ec='b')
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x, p(x),color='red')
```

```python
# Task 7
x=data[data.Department=="IT "]
y=data[data.Department=="Sales"]
z=data[data.Department=="Finance"]
w=data[data.Department=="HR"]
```

```python
#ploting box blot for all departments and using subplot to proper visualization
plt.subplot(2,2,1)
plt.boxplot(x.iloc[:,3])
plt.title("IT")

plt.subplot(2,2,2)
plt.boxplot(y.iloc[:,3])
plt.title("Sales")

plt.subplot(2,2,3)
plt.boxplot(z.iloc[:,3])
plt.title("Finance")

plt.subplot(2,2,4)
plt.boxplot(w.iloc[:,3])
plt.title("HR")
```
